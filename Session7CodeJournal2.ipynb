```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)
```

```python
def exp(x):
    return np.exp(x)
```

```python
y = exp(x)
```

```python
f, ax = plt.subplots(1,1, figsize=(4,4))
ax.plot(x,y,color='green',linewidth=1.5)
ax.set_xlabel('Time [milliseconds]',fontsize=14)
ax.set_ylabel('Awesomeness',fontsize=14)
plt.savefig('Session7plot.pdf',bbox_inches='tight',dpi=400)
```
