```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.poisson(lam=1,size=1000)
```

```python
f, ax = plt.subplots(1,1,figsize=(4,4))

ax.hist(x,facecolor='orange',bins=100,edgecolor='black',alpha=0.5)

ax.set_xlabel('Poisson Distribution',fontsize=14)
ax.set_ylabel('Frequency',fontsize=14)

plt.savefig('Session10hist.pdf',bbox_inches='tight',dpi=400)
```
