```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)
```

```python
def sin(x):
    return np.sin(x)
def cos(x):
    return np.cos(x)
```

```python
sin = sin(x)
cos = cos(x)
```

```python
f, ax = plt.subplots(1,2, figsize=(8,4))
ax[0].plot(x,sin,'red',linewidth=1.5,label='model (line)')
ax[0].set_xlabel('x',fontsize=14)
ax[0].set_ylabel('sin(x)',fontsize=14)

ax[1].plot(x,cos,'blue',linewidth=1.5,label='model (line)')
ax[1].set_xlabel('x',fontsize=14)
ax[1].set_ylabel('cos(x)',fontsize=14)

plt.tight_layout()

plt.savefig('Session8plot.pdf',bbox_inches='tight',dpi=400)
```
